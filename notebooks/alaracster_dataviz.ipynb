```python
df['hora_creacion'] = pd.to_timedelta(df['hora_creacion'])
```

```python
df['hora_cierre'] = pd.to_timedelta(df['hora_cierre'])
df.head()
```

```python
df['hora_creacion'] = df['hora_creacion'].dt.total_seconds() / (60 * 60)
```

```python
df['hora_cierre'] = df['hora_cierre'].dt.total_seconds() / (60 * 60)
```

```python

df.head()
```

```python

imputer = SimpleImputer(strategy='most_frequent')
```

```python
imputer.fit(df[['hora_creacion']])
df['hora_creacion'] = imputer.transform(df[['hora_creacion']])
imputer.statistics_
```

```python
imputer.fit(df[['hora_cierre']])
df['hora_cierre'] = imputer.transform(df[['hora_cierre']])
imputer.statistics_
```

```python
df['hora_creacion'] = df['hora_creacion'] *60*60
df['hora_cierre'] = df['hora_cierre'] *60*60
```

```python
df['hora_creacion'] = pd.to_timedelta(df['hora_creacion'], unit='s')
df['hora_cierre'] = pd.to_timedelta(df['hora_cierre'], unit='s')
```

```python
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['fecha_creacion'] = pd.to_datetime(df['fecha_creacion'])
df.dtypes
```

```python
def is_rush_hour(row):
    hour_calc = row['hora_creacion'].seconds // 3600  # Convertir el timedelta a horas

    if hour_calc >= 7.5 and hour_calc <= 9.5:  # Verificar la hora entre 07:30 y 09:30
        return 1
    elif row['fecha_creacion'].year >= 2021 and (hour_calc >= 14 and hour_calc <= 15):  # A partir de marzo de 2020, verificar la hora entre 14:00 y 15:00
        return 1
    elif hour_calc >= 18 and hour_calc <= 19:  # Verificar la hora entre 18:00 y 19:00
        return 1
    else:
        return 0

# Aplicar la función a la columna 'hora_creacion' y almacenar los resultados en una nueva columna 'is_rush_hour'
df['is_rush_hour'] = df.apply(is_rush_hour, axis=1)
```
